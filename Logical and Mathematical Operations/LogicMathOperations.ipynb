### **Operaciones logicas**
En primer lugar vamos a crear dos imagenes negras de 300x300, una llamada image0 y la otra image1, con un rectangulo blanco en cada una de ellas en las siguientes posiciones y tama√±os.

```
# Para la imagen0
p0 = (25,100) 
size0 = (200,250)

#Para la iamgen1
p1 = (150,50)
size1 = (250,150)
```
Recordar como crear imagenes negras [aqui](http://www.learningaboutelectronics.com/Articles/How-to-create-a-blank-image-in-Python-numpy.php)

Recordar como pintar cuadrados [aqui](https://www.geeksforgeeks.org/python-opencv-cv2-rectangle-method/)


````python
image_size = (300,300)
image0 = np.zeros(image_size, dtype='uint8')
image1 = np.zeros(image_size, dtype='uint8')

cv2.rectangle(image0, (25,100), (200,250), 255, -1)
cv2.rectangle(image1, (150,50), (250,150), 255, -1)
plt.subplot(121), plt.imshow(image0, cmap = 'gray')
plt.subplot(122), plt.imshow(image1, cmap = 'gray')
plt.show()
````

Para la realizacion de las siguientes actividades podemos consultar la doc de opencv de operaciones aritmeticas [aqui](https://docs.opencv.org/4.x/d0/d86/tutorial_py_image_arithmetics.html)

### **Bitwise not**
Invertir aritmeticamente la imagen0

````python
output_image = cv2.bitwise_not(image0)
plt.imshow(output_image, cmap = 'gray')
plt.show()
````

### **Bitwise and**
Hacer una and entre la imagen0 y la imagen1

````python
put_image = cv2.bitwise_and(image0, image1)
plt.imshow(output_image, cmap = 'gray')
plt.show()
````

### **Bitwise or**
Hacer una operacion or entre la imagen0 y la imagen1

````python
output_image = cv2.bitwise_or(image0, image1)
plt.imshow(output_image, cmap = 'gray')
plt.show()
````

### **Bitwise xor**
Hacer una operacion xor entre la imagen0 y la imagen1

````python
output_image = cv2.bitwise_xor(image0, image1)
plt.imshow(output_image, cmap = 'gray')
plt.show()
````

### **Operaciones aritmeticas**
A continuacion cargaremos las siguientes imagenes para practicar operaciones aritmeticas

````python
# nombre de la imagen a cargar
image_path_0= 'data/bugs_bunny.jpg'
image_path_1= 'data/daffy_duck.jpg'

# cargamos la imagen
bugs_bunny = cv2.cvtColor(cv2.imread(image_path_0), cv2.COLOR_RGB2BGR)
duffy_duck = cv2.cvtColor(cv2.imread(image_path_1), cv2.COLOR_RGB2BGR)
plt.subplot(121), plt.imshow(duffy_duck)
plt.subplot(122), plt.imshow(bugs_bunny)
plt.show()
````

### **Suma**
Para sumar imagenes podemos usar dos metodos, suma de vectores o el metodo cv2.add()
En el siguiente codigo podemos ver la diferencia de usar uno y el otro. En una suma de vecotres, cuando ocurre overflow el valor de pixel se reinicia mientras que con el metodo add tenemos un clipping entre 0 y 255 que tiene mas sentido al trabajar con imagenes

````python
x0 = np.uint8([250])
x1 = np.uint8([10])

print(x0 + x1)
print(cv2.add(x0,x1))
````

Sumar con los dos metodos anteriores las imagenes del pato Lucas y Bugs Bunny. Mostrar un subtplot con los dos resultados

````python
output_image_0 = (bugs_bunny + duffy_duck).astype(np.uint8)
output_image_1 = cv2.add(bugs_bunny, duffy_duck)

plt.subplot(121), plt.imshow(output_image_0)
plt.subplot(122), plt.imshow(output_image_1)
plt.show()
````

### **Resta**
Con la resta nos pasa exacmatente lo mismo que con la operacion de suma


````python
x0 = np.uint8([0])
x1 = np.uint8([10])

print(x0 - x1)
print(cv2.subtract(x0,x1))
````

Restas con los dos metodos anteriores las imagenes del pato Lucas y Bugs Bunny. Mostrar un subtplot con los dos resultados

````python
output_image_0 = (bugs_bunny - duffy_duck).astype(np.uint8)
output_image_1 = cv2.subtract(bugs_bunny, duffy_duck)

plt.subplot(121), plt.imshow(output_image_0)
plt.subplot(122), plt.imshow(output_image_1)
plt.show()
````

### **Suma ponderada**
Realizar una suma ponderada de las dos imagenes usando el metodo cv2.addWeighted()


````python
output_image = cv2.addWeighted(bugs_bunny,0.7,duffy_duck,0.3,0)
plt.imshow(output_image)
plt.show()
````

### **Multiplicacion**
Con la multiplicacion nos pasa exacmatente lo mismo que con la operacion de suma


````python
x0 = np.uint8([129])
x1 = np.uint8([2])

print(x0 * x1)
print(cv2.multiply(x0,x1))
````

Multiplicar cada canal de las imagenes por una constante.
Mas info del metodo multiply [aqui](https://docs.opencv.org/3.4/d2/de8/group__core__array.html#ga979d898a58d7f61c53003e162e7ad89f)

````python
output_image_0 = cv2.multiply(duffy_duck, (4,1,3,2))
output_image_1 = cv2.multiply(bugs_bunny, (2,3,1,2))

plt.subplot(121), plt.imshow(output_image_0)
plt.subplot(122), plt.imshow(output_image_1)
plt.show()
````

### **Divisi√≥n**
Dividir la imagen de bugs bunny entre 2 para cada canal


````python
output_image_0 = (bugs_bunny / 2).astype(np.uint8)
output_image_1 = cv2.divide(bugs_bunny, (2,2,2,0))

plt.subplot(121), plt.imshow(output_image_0)
plt.subplot(122), plt.imshow(output_image_1)
plt.show()
````

### **Maximo y minimo**
Obtener la imagen de maximo y minimo entre las dos imagenes usando


```
max = cv2.max(img0, img1)
min = cv2.min(img0, img1)
```




````python
output_image_0 = cv2.max(duffy_duck,bugs_bunny)
output_image_1 = cv2.min(duffy_duck,bugs_bunny)
plt.subplot(121), plt.imshow(output_image_0)
plt.subplot(122), plt.imshow(output_image_1)
plt.show()
````

### **Diferencia absoluta**
Obtener la imagen de absoulta differencia

````python
# output_image = cv2.absdiff(duffy_duck,bugs_bunny)
output_image = cv2.absdiff(cv2.cvtColor(duffy_duck, cv2.COLOR_BGR2GRAY),
                               cv2.cvtColor(bugs_bunny, cv2.COLOR_BGR2GRAY))
plt.imshow(output_image)
plt.show()
````
