**Metodo para printar los datos de un np_array**

Utilizaremos este metodo de aqui en adelenta para poder ver la información del array

````python
def array_info(array):
    print('Array:\n{}'.format(array))
    print('Data type:\t{}'.format(array.dtype))
    print('Array shape:\t{}\n'.format(array.shape))


# ejemplo de uso
a = np.array([0,1,2])

array_info(a)
````

**Pintar con palabras clave**

Hacer el plot de puntos los siguientes valores:



```
data = {'a': np.arange(25),
        'c': np.random.randint(0, 25, 25),
        'd': np.random.randn(25)}
data['b'] = data['a'] + 10 * np.random.randn(25)
data['d'] = np.abs(data['d']) * 100
```



````python
data = {'a': np.arange(25),
        'c': np.random.randint(0, 25, 25),
        'd': np.random.randn(25)}
data['b'] = data['a'] + 10 * np.random.randn(25)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

**Subplot**

Hacer tres graficosde los siguientes valores, el primero formato bar, el segundo scatter y el tercero linea con plot:


```
names = ['group_a', 'group_b', 'group_c']
values = [-50, 100, 25]
```

````python
names = ['group_a', 'group_b', 'group_c']
values = [-50, 100, 25]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````

````python
# ruta de la imagen a utilizar
image_path = 'data/cat.jpg'
````

### **Scikit-image**

Usar scikit-image para abrir la image y matplotlib para mostrarla


````python
cat = io.imread(image_path)
io.imshow(cat)
io.show()
````

### **Opencv**

Usar opencv para abrir la image y matplotlib para mostrarla


````python
cat = cv2.imread(image_path)
im_rgb  = cv2.cvtColor(cat, cv2.COLOR_RGB2BGR)
plt.imshow(im_rgb)
plt.show()
````

### **Pillow**

Usar Pillow para abrir la image y matplotlib para mostrarla

````python
cat = Image.open(image_path)
plt.imshow(cat)
plt.show()
````
